1. Muestra un gráfico de barras con el total de productos por categoria
2. Muestra un gráfico circular con el precio medio por categoria
3. Muestra un histograma con los precios de los productos

```python
import pandas as pd
import matplotlib.pyplot as plt

datos = pd.read_csv("products.csv")
df = datos[['productID', 'productName', 'supplierID', 'categoryID', 'quantityPerUnit', 'unitPrice', 'unitsInStock', 'unitsOnOrder', 'reorderLevel']]
print(datos)

# gráfico de barras con el total de clientes por país
total_productos = df.groupby('productName')['categoryID'].mean()
total_productos.head(10).plot.barh()
plt.show()


# gráfico circular con el precio medio por categoría
df.unitPrice.value_counts().head(10).plot.pie()
plt.show()


# histograma precio productos
df.unitPrice.plot.hist()
plt.show()
```
